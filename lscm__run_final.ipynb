```python
import pandas as pd
df = pd.read_csv('roa.csv')
```

```python
df
```

```python
df['inv'] = df['inv'].str.replace(',', '')
df['inv'] = df['inv'].astype(int)
df['size'] = df['size'].str.replace(',', '')
df['size'] = df['size'].astype(int)
```

```python
df.dtypes
```

```python
df.columns
```

```python
X = df.drop('roa', axis = 1)
y = df['roa']
```

```python
X
```

```python
from sklearn.linear_model import Lasso
from sklearn.preprocessing import MinMaxScaler
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
import numpy as np

pipeline = Pipeline([('scaler',MinMaxScaler()), ('model',Lasso())])
search = GridSearchCV(pipeline, {'model__alpha':np.arange(0.1,10000,0.1)},cv = 5, scoring="neg_mean_squared_error",verbose=3)
search.fit(X,y)
alpha = search.best_params_
coefficients = search.best_estimator_.named_steps['model'].coef_
importance = np.abs(coefficients)
```

```python
val = -1
pos = 0

for i in range(len(importance)):
  if val < importance[i]:
    val = importance[i]
    pos = i


print(np.max(importance))
print(importance)
print(val)
print(pos + 1)
```
